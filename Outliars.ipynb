### Feature is skewed

```python
fig=df.Fare.hist(bins=50)
fig.set_title("Fare")
fig.set_xlabel('Fare')
fig.set_xlabel('no_passengers')

## sweked dataset ka mtln not normally distributed
```

```python
figure=df.boxplot(column="Fare")
```

calculate interquantile range to calculate boundries

hm log yeh assume krte hai ki 75 and 25 percentile ke bhar sb outliars hoga isliye apn log Iqr lete hai

IQR=75%-25%(difference between 75 percentile and 25 percentile)

```python
IQR=df['Fare'].quantile(0.75)-df['Fare'].quantile(0.25)
IQR
```

```python
lowerBridge=df['Fare'].quantile(0.25)-IQR*1.5
upperBridge=df['Fare'].quantile(0.75)+IQR*1.5
print(lowerBridge),print(upperBridge)
```

```python
### extreme outliars
lowerBridge=df['Fare'].quantile(0.25)-IQR*3
upperBridge=df['Fare'].quantile(0.75)+IQR*3
print(lowerBridge),print(upperBridge)
```
